```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('student_performance.csv')

# a) Display the dataset dimensions
print(f"Shape: {df.shape[0]} students and {df.shape[1]} columns")

# b) Inspect column names and data types
print("\n--- Column Types ---")
print(df.dtypes)

# c) Display the first 10 rows to verify data loading
print("\n--- First 10 Rows ---")
display(df.head(10))

# d) Check for missing values across all columns
print("\n--- Missing Values ---")
print(df.isnull().sum())

# e) Analyze the distribution of the target variable 'status'
print("\n--- Status Distribution ---")
# Check distribution of Dropout, Graduate, and Enrolled students
print(df['status'].value_counts())
```

1.2 Program Performance Dashboard

```python
# 1.2 Program Performance Dashboard

# Define a custom aggregation function to count dropouts
def dropout_count(x):
    return (x == 'Dropout').sum()

# Aggregate data by Course to calculate key performance metrics
# Columns selected: Status, Admission Grade, Sem 1 Grade, Sem 2 Grade
program_stats = df.groupby('course').agg({
    'status': ['count', dropout_count],
    'admission_grade': 'mean',
    'semester1_grade': 'mean',
    'semester2_grade': 'mean'
})

# Flatten the MultiIndex columns and rename for clarity
program_stats.columns = [
    'Total Students',
    'Dropout Count',
    'Avg Admission Grade',
    'Avg Sem 1 Grade',
    'Avg Sem 2 Grade'
]

# Calculate the Dropout Rate percentage for each program
program_stats['Dropout Rate (%)'] = (program_stats['Dropout Count'] / program_stats['Total Students']) * 100

# Sort by Dropout Rate descending to identify high-risk programs
display(program_stats.sort_values(by='Dropout Rate (%)', ascending=False))
```

1.2 Program Analysis

Highest & Lowest Dropout Rates: The program with the highest dropout rate is Biofuel Production (66.7%), followed closely by Equinculture (55.3%) and Informatics Engineering (54.1%). The program with the lowest dropout rate is Nursing (15.4%).

* Admission Grades vs. Retention: Programs with higher admission grades do not consistently have lower dropout rates. for example, Equinculture has a high average admission grade (approx. 132.4) yet maintains one of the highest dropout rates, suggesting that pre-admission academic performance is not the sole predictor of retention.

* Hypothesis: The high dropout rate in Informatics Engineering (54.1%) despite reasonable admission grades may be due to the rigorous nature of the curriculum or a "shock" in the first semester, as evidenced by their low average Semester 1 grade (8.56), which is the second lowest among all programs.

* Recommendation: For the Informatics Engineering program, I recommend implementing mandatory supplemental instruction or peer tutoring for first-year courses to improve Semester 1 performance, as early academic struggles appear to be a key driver of attrition in this major.

PROBLEM 2: Identifying Risk Factors


2.1 Academic Performance Patterns

```python
# 2.1 Academic Performance Patterns

# --- Part A: Metrics by Status ---

# Calculate Semester 1 Success Rate (Approved / Enrolled)
# We use np.where to handle division by zero (if a student enrolled in 0 courses)
df['sem1_success_pct'] = np.where(df['semester1_enrolled'] > 0,
                                  df['semester1_approved'] / df['semester1_enrolled'],
                                  0)

# Identify students who failed ALL semester 1 courses (0 approved courses)
# This is a binary flag (1 = Failed All, 0 = Did not fail all)
df['failed_all_sem1'] = np.where(df['semester1_approved'] == 0, 1, 0)

# Group by Status to see patterns
# We calculate the mean to get averages and percentages
status_metrics = df.groupby('status').agg({
    'sem1_success_pct': 'mean',          # Average success rate
    'semester1_grade': 'mean',           # Average grade
    'failed_all_sem1': 'mean'            # % who failed all (mean of 0s and 1s = %)
})

print("--- Part A: Academic Patterns by Status ---")
display(status_metrics)


# --- Part B: At-Risk Analysis (< 50% Success) ---

# Define At-Risk: Success rate < 0.50 (50%)
df['at_risk'] = df['sem1_success_pct'] < 0.50

# 1. Percentage of all students who are at-risk
pct_at_risk = df['at_risk'].mean() * 100

# 2. Dropout rate among At-Risk students
# Filter for at-risk, then check dropout rate
at_risk_group = df[df['at_risk'] == True]
dropout_rate_risk = (at_risk_group['status'] == 'Dropout').mean() * 100

# 3. Dropout rate among NOT At-Risk students
safe_group = df[df['at_risk'] == False]
dropout_rate_safe = (safe_group['status'] == 'Dropout').mean() * 100

print("\n--- Part B: At-Risk Analysis ---")
print(f"Percentage of students considered 'At-Risk': {pct_at_risk:.2f}%")
print(f"Dropout Rate (At-Risk Group): {dropout_rate_risk:.2f}%")
print(f"Dropout Rate (Not At-Risk Group): {dropout_rate_safe:.2f}%")
```

2.1 Academic Patterns & Risk Analysis

Predictive Power: Semester 1 performance is a definitive predictor of dropout. The data shows that students who dropout have a drastically lower success rate in their first semester compared to graduates.

Threshold Effectiveness: The 50% success rate threshold is highly effective. Students identified as "At-Risk" (passing fewer than half their classes) have a dropout rate of 77.4%, whereas students who pass more than half their classes have a dropout rate of only 18.4%. This clearly separates the high-risk population.

Timing of Intervention: Intervention must occur immediately after Semester 1. Waiting for Semester 2 is too late, as 40.1% of all dropouts failed every single course in their first semester, indicating they have already fully disengaged from the academic program.

Recommended Action: The university should automatically place any student with a Sem 1 success rate below 50% into a mandatory "Academic Probation & Recovery" program. This program would require bi-weekly meetings with an academic advisor to build a manageable course load for Semester 2.

Problem 2.2: Demographics and Financial Factors

```python
# 2.2 Demographics and Financial Factors

# --- Part A: Impact of Scholarships ---
# Group by Scholarship (Yes/No)
scholarship_stats = df.groupby('scholarship').agg({
    'status': lambda x: (x == 'Dropout').mean() * 100, # Calculate Dropout Rate %
    'semester1_grade': 'mean',
    'semester2_grade': 'mean'
})
scholarship_stats.columns = ['Dropout Rate (%)', 'Avg Sem 1 Grade', 'Avg Sem 2 Grade']

print("--- Part A: Impact of Scholarships ---")
display(scholarship_stats)


# --- Part B: Impact of Age ---
# Create Age Bins: Traditional (17-22), Young Adult (23-30), Mature (31+)
age_bins = [0, 22, 30, 100]
age_labels = ['Traditional (17-22)', 'Young Adult (23-30)', 'Mature (31+)']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)

# Group by Age Group
# Note: observed=False handles categorical data warnings in newer pandas versions
age_stats = df.groupby('age_group', observed=False).agg({
    'status': ['count', lambda x: (x == 'Dropout').mean() * 100],
    'semester1_grade': 'mean'
})
age_stats.columns = ['Student Count', 'Dropout Rate (%)', 'Avg Sem 1 Grade']

print("\n--- Part B: Impact of Age ---")
display(age_stats)


# --- Part C: Financial Stability ---
# Define Financial Issues: Debtor ('Yes') OR Tuition Up to Date ('No')
# We use logical OR (|) to flag students with EITHER issue
df['financial_issues'] = (df['debtor'] == 'Yes') | (df['tuition_up_to_date'] == 'No')

financial_stats = df.groupby('financial_issues').agg({
    'status': lambda x: (x == 'Dropout').mean() * 100
})
financial_stats.columns = ['Dropout Rate (%)']
# Rename index for clarity
financial_stats.index = ['No Financial Issues', 'Has Financial Issues']

print("\n--- Part C: Financial Stability ---")
display(financial_stats)
```

2.2 Demographics and Financial Analysis:

Scholarships: Scholarships significantly improve retention. Students with scholarships have a dropout rate of 12.2%, which is roughly one-third of the rate for students without scholarships (38.7%). This suggests that financial support (and potentially the academic requirement to maintain it) is a strong protective factor.

Age Groups: The Mature (31+) age group has the highest dropout risk at 53.6%, followed closely by Young Adults at 52.2%. In contrast, Traditional students (17-22) perform much better with a dropout rate of only 22.7%, likely due to fewer external responsibilities such as full-time employment or family care.

Financial Status: Financial status is a critical determinant of success. Students flagged with financial issues (tuition debt or late payments) have a staggering dropout rate of 70.6%, compared to just 23.8% for those in good financial standing.

Recommendation: My #1 recommendation is to implement an Emergency Micro-Grant Program specifically targeting the students flagged with "Financial Issues." Since over 70% of this group drops out, resolving small tuition arrears or providing short-term financial relief could yield the highest immediate return on retention.
